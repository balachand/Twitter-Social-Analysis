```python
data = pd.read_csv("C:\\Bala\\DataScience\\Twitter\\Followers-BFussien.csv")
```

```python
data['influencer'] = 'BFussien'
```

```python
writeDf = data[['influencer' , 'Screen Name']]
writeDf.to_csv("C:\\Bala\\DataScience\\Twitter\\Follower-Graph.csv",index=False)
```

```python
df = writeDf
o = net.DiGraph()
df.apply(lambda row : o.add_edge(row['influencer'] , row['Screen Name'] , color='red'),axis=1)

SEED = 'BFussien'
g = net.DiGraph(net.ego_graph(o,SEED,radius=2))

plt.figure(figsize=(18,18))
plt.axis('off')
pos = net.spring_layout(o)
colours = ['red','green']
colourmap = {}

net.draw_networkx_nodes(o, pos, node_color='green')
net.draw_networkx_edges(o, pos, width=0.3, alpha=0.1)
net.draw_networkx_labels(o, pos)

plt.show()
```
